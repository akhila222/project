```python
sns.pairplot(data)
```

```python
sns.set(rc={"figure.figsize":(12,6)})
sns.countplot(x='job',data=data)
plt.title('job distribution')
plt.xticks(rotation=90)
plt.yticks()
plt.show()
```

```python
sns.set(rc={"figure.figsize":(10,5)})
sns.countplot(x='education',data=data)
plt.title('Education of customers')
plt.xticks(rotation=90)
plt.yticks()
plt.show()
```

```python
plt.figure()
plt.hist(data['age'],color='purple')
plt.title('Age distribution of customers')
plt.xticks()
plt.yticks()
```

```python
df=pd.DataFrame(data)
housing=df['housing'].value_counts()
housing.plot(kind='bar',figsize=(6,5))
plt.title('housing loan of customers')
plt.show()
```

```python
df=pd.DataFrame(data)
loan=df['loan'].value_counts()
loan.plot(kind='bar',figsize=(6,5),color='blue')
plt.title('personal loan')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='campaign',y='subscribed',data=data)
plt.show()
```

```python
sns.jointplot(x=data["campaign"], y=data["subscribed"], kind='scatter')
```

```python
plt.figure()
sns.histplot(data=data,x="age",hue="loan",bins=30)
plt.title("Age Distribution of Customers vs their Loan status")
plt.show()
```
